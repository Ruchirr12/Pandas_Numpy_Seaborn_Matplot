# LABELLING

```python
plt.figure(figsize=(13,4))
plt.hist(data["mcg"],color="aquamarine",edgecolor="teal")
plt.title("MCG HISTOGRAM")
plt.xlabel("mcg")
plt.ylabel("frequency")
plt.axis("tight")
plt.show()
plt.legend()
```

# Customisation of histogram

```python
plt.hist(data.alm,bins=20,edgecolor="r",color="g",histtype="stepfilled",alpha=.9)
```

# Histogram of different columns on one graph

```python
plt.figure(figsize=(18,4))
plt.hist(data.alm,label="alm")
plt.hist(data.mcg,label="mcg")
plt.hist(data.gvh,label="gvh")
plt.legend()
```

```python
data.head()
```

```python
wer=data.select_dtypes("object")
wer
```

```python
wer1=data.drop(wer,axis=1)
wer1
```

```python
b=plt.figure("figsize",(18,4))
a=dict(bins=15,edgecolor="g",color="r")
for i in wer1:
    plt.figure(figsize=(11,4))
    plt.hist(data[i],**a)
    plt.title("histogram of {}".format(i))
    plt.xlabel("{}".format(i))
    plt.ylabel("frequency")
    plt.show()
    
    
   
```

```python
b=plt.figure(figsize=(25,10))
col=["r","y","c","olive","orange","peru","b","g"]
index=1
for i in wer1:
        ax=plt.subplot(2,4,index)
        plt.hist(data[i],label=i,bins=20,color=col[index-1],edgecolor="k")
        plt.title("HISTOGRAM OF {}".format(i))
        plt.xlabel(i)
        plt.ylabel("FREQUENCY OF {}".format(i))
        plt.legend(loc="upper left")
        ax.axis("tight")
        index+=1
```
