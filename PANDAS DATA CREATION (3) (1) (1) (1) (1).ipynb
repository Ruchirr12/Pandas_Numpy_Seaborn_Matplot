# MERGING BY COMMON KEY

```python
a = pd.DataFrame({'order_id': ['K0', 'K9', 'K2', 'K3'],
                     'temp': ['A0', 'A1', 'A2', 'A3'],
                     'windspeed': ['B0', 'B1', 'B2', 'B3']})
   
b = pd.DataFrame({'order_id': ['K0', 'K1', 'K2', 'K3'],
                          'humidity': ['C0', 'C1', 'C2', 'C3'],
                          'event': ['D0', 'D1', 'D2', 'D3']})

c= pd.DataFrame({'order_id': ['K0', 'K8', 'K2', 'K3'],
                          'country': ['C0', 'C1', 'C2', 'C3'],
                          'state': ['D0', 'D1', 'D2', 'D3']})
```

```python
a
```

```python
b
```

```python
pd.merge(a,b,how="left")
```

```python
merge1=pd.merge(a,b,how="outer")
```

```python
merge2=pd.merge(merge1,c,how="outer")
```

```python
merge2
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K1', 'K0', 'K1', 'K2'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K1', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
merge2=pd.merge(left,right,how ="outer")
```

```python
merge2
```

```python
left,right
```

```python
merge3=pd.merge(left,right,how="inner",on=["key1"])
```

```python
merge3
```

```python
left
```

```python
merge5=pd.merge(left,right,how="outer",on=["key1","key2"])
```

```python
merge5
```

# JOINTS

```python
a = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']}
                     ,index=[0,1,2] )

b = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},index=[0,2,3]
                     )
```

```python
a
```

```python
b
```

```python
a.join(b,how="right")
```

```python
b.join(a,how="inner")
```

```python
a.join(b,how="inner")
```

# EXAMPLES

```python
import pandas as pd
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["orlando","chicago","texas"],
    "temperature": [32,14,35],
    "humidity": [80, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather],axis=0)
df
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
})
windspeed_df
```

```python
pd.merge(temperature_df,windspeed_df,"outer")
```
